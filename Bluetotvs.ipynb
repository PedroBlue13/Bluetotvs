```python
import os
import sys
import time
import tkinter as tk
from tkinter import messagebox

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

payload = {
    "userauth": "49513449807",
    "passauth": "000000"
}





chrome_options = webdriver.ChromeOptions()
# chrome_options.add_argument("--window-position=-10000,-10000")  # Modo off-screen
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")
chrome_options.add_argument("--kiosk-printing") 

prefs = {
            "printing.print_preview_sticky_settings.appState": '{"recentDestinations":[{"id":"Save as PDF","origin":"local","account":""}],"selectedDestinationId":"Save as PDF","version":2}',          
            "download.prompt_for_download": False,  
            "download.directory_upgrade": True
        }


chrome_options.add_experimental_option("prefs", prefs)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
print (f'acessando site')




driver.get("https://helisultaxi119564.protheus.cloudtotvs.com.br:4020/meurh01/#/login")
time.sleep(4)
print('carregou')

element_selector_user = "body > app-root > div > div > app-login > div.po-sm-12.po-md-5.po-lg-5.po-xl-5.login-container > div.po-sm-12.po-md-5.po-lg-5.po-xl-5.container-normal-login.login-form > form > po-input > po-field-container > div > div.po-field-container-content > input"
element_selector_pass = "body > app-root > div > div > app-login > div.po-sm-12.po-md-5.po-lg-5.po-xl-5.login-container > div.po-sm-12.po-md-5.po-lg-5.po-xl-5.container-normal-login.login-form > form > po-password > po-field-container > div > div.po-field-container-content > input"

max_try=5
for attempt in range(max_try):
    try:
        
        user_input = driver.find_element(By.CSS_SELECTOR, element_selector_user)

        
        user_input.clear() 
        for char in payload["userauth"]:
            user_input.send_keys(char) 
            time.sleep(0.1)  

        # Validar se o valor foi digitado corretamente
        if user_input.get_attribute("value") == payload["userauth"]:
            print(f"Usuário {payload['userauth']} digitado com sucesso!")
            break
    except Exception as e:
        print(f"Tentativa falha: {e}")
        time.sleep(2)
else:
    print("Não foi possível digitar o usuário. Tente novamente mais tarde!")

time.sleep(2)

for attempt in range(max_try):
    try:
        pass_input = driver.find_element(By.CSS_SELECTOR, element_selector_pass)

        
        pass_input.clear()  
        for char in payload["passauth"]:
            pass_input.send_keys(char)  
            time.sleep(0.1)  

  
        if pass_input.get_attribute("value") == payload["passauth"]:
            print(f"Senha {payload['passauth']} digitada com sucesso!")
            break
    except Exception as e:
        print(f"Tentativa falha: {e}")
        time.sleep(2)
else:
    print("Não foi possível digitar a senha. Tente novamente mais tarde!")

time.sleep(1)

button_login = "/html/body/app-root/div/div/app-login/div[1]/div[1]/form/po-button[1]/button"
try:
    button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, button_login))
    )
    button.click()
    print("Entramos!")
except Exception as e:
    print(f'Não conseguimos por: {e}')

button_point = "/html/body/app-root/div/div/div[2]/po-menu/div[2]/nav/div/div/div[3]/po-menu-item/div/div[1]"
try:
    button = WebDriverWait(driver, 20).until(
         EC.element_to_be_clickable((By.XPATH, button_point))
    )
    button.click()
    print("Abrimos o dropdown")
except Exception as e:
    print("Não abrimos o dropdown: {e}")

button_setpoint = "/html/body/app-root/div/div/div[2]/po-menu/div[2]/nav/div/div/div[3]/po-menu-item/div/div[2]/div[3]/po-menu-item/a/div"
try:
    button = WebDriverWait(driver, 30).until(
         EC.element_to_be_clickable((By.XPATH, button_setpoint))
    )
    button.click()
    print("Abrimos o ponto")
except Exception as e:
    print("Não abrimos o ponto: {e}")


button_hitpoint = '//*[@id="btn-app-swipe-clocking-register"]/div/div/div/div'
try:
    button = WebDriverWait(driver, 40).until(
         EC.element_to_be_clickable((By.XPATH, button_hitpoint))
    )
    button.click()
    print("Batemos o ponto")
except Exception as e:
    print(f"Não Batemos o ponto: {e}")
```
